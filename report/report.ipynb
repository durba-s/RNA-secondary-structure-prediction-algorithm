```python
import numpy as np
from sklearn.linear_model import LinearRegression
X = np.array(data['n']**3).reshape(-1, 1)
y = np.array(data['dp'])
reg = LinearRegression().fit(X, y)
reg.score(X, y)
```

```python
bases = ['A','U','C','G']
n = [5,10,15,20,25,30,40,50,70,100,125,150,175,200,250,300,250,400,450,500,550,600,700,800,900]
rna = []
for i in range(len(n)):
    seqlen = n[i]
    seq = ''
    for l in range(seqlen):
        seq = seq+np.random.choice(bases)
    rna.append(seq)
```

```python
with open('eg.txt', 'w') as f:
    for item in rna:
        f.write("%s\n" % item)
```
